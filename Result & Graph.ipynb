```python
import matplotlib.pyplot as plt
  
x = [200,500,1000,100,50,10]
x.sort()
x1=[871.98, 2440.59, 4015.17, 387.46, 212.47, 45.68]
x1.sort()
plt.plot(x, x1,label="process")
plt.xlabel('No.of images')
plt.ylabel('Time in sec')
plt.title('Regular compression')
plt.show()
```

```python
import matplotlib.pyplot as plt
  
x = [200,500,1000,100,50,10]
x2=[730.82, 1835.92, 3668.36, 370.72, 190.21, 40.76]
x1=[871.98, 2440.59, 4015.17, 387.46, 212.47, 45.68]
x.sort()
x1.sort()
x2.sort()
plt.plot(x, x1,label="1 process")
plt.plot(x, x2,label="2 process")
plt.xlabel('No.of images')
plt.ylabel('Time in sec')
plt.legend()
plt.title('Compare between 1 processor and 2 processors')
plt.show()
```
