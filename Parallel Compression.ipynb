```python
import time
import numpy as np
import math
import cv2
import glob
import matplotlib.pyplot as plt
start=time.time()
image_array = []
files = glob.glob (r"C:\Users\charan ps\Downloads\cats10\cat.*.jpg")
for myFile in files:
    image = cv2.imread (myFile)
    image_array.append (image)
print('image_array shape:', np.array(image_array).shape) 
```

```python
import multiprocessing
import Untitled8
pool = multiprocessing.Pool()
pool = multiprocessing.Pool(processes=2)
pool.map(Untitled8.hpc1,image_array)
timee.append(time.time()-start)
```

```python
print(timee)
```

```python
x = [200,500,1000,100,50,10]
  
plt.plot(x, timee,label="line1")
plt.xlabel('No.of images')
plt.ylabel('Time(sec)')
  
# giving a title to my graph
plt.title('Parallel with 2 processors')
  
# function to show the plot
plt.show()
```
